## Method to calculate an exact integral sybolically using Gauss-Legendre quadrature

```python
import sympy as sp

sp.init_printing()
x = sp.Symbol("x")


def gauss_legendre(n):
    x = sp.Symbol("x")
    Pnx = sp.legendre(n, x)
    Pp = sp.diff(Pnx, x)
    xi = sp.solve(Pnx, x)
    wi = [sp.simplify(2 / (1 - xj**2) / (Pp.subs(x, xj)) ** 2) for xj in xi]
    return xi, wi


xs, ws = gauss_legendre(2)

print(xs)
print(ws)
```

```python
p_roots = gauss_legendre


def integrate_gauss(expr, x, n=1, dx=1):
    x_norm, w_norm = p_roots(n)

    ws = [w_ * dx / 2 for w_ in w_norm]
    xs = [dx / 2 * (1 + x_) for x_ in x_norm]

    # gauss legendre calculation
    s = 0
    for xi, wi in zip(xs, ws):
        s += wi * expr.subs(x, xi)
    return s


import sympy

x = sympy.Symbol("x")
a = sympy.Symbol("a")
sympy.simplify(integrate_gauss(a, x, n=3))
```

```python
def test_integrate_gauss():
    import sympy

    x = sympy.Symbol("x")
    a = sympy.Symbol("a")

    assert sympy.simplify(integrate_gauss(a, x, n=1)) == a
    assert sympy.simplify(integrate_gauss(a, x, n=2)) == a
    assert sympy.simplify(integrate_gauss(a, x, n=3)) == a


test_integrate_gauss()
```

```python
def integrate_gauss_cuboid(expr, x, y, z, dx, dy, dz, n=2):
    s1 = integrate_gauss(expr, x=x, n=n, dx=dx)
    s2 = integrate_gauss(s1, x=y, n=n, dx=dy)
    s3 = integrate_gauss(s2, x=z, n=n, dx=dz)
    return s3


def test_integrate_gauss_cuboid():
    x, y, z, dx, dy, dz = sympy.symbols(["x", "y", "z", "dx", "dy", "dz"])
    assert (
        sympy.simplify(integrate_gauss_cuboid(a, x, y, dz, dx, dy, dz))
        == a * dx * dy * dz
    )
    assert (
        sympy.simplify(integrate_gauss_cuboid(a * x, x, y, dz, dx, dy, dz))
        == a * dx**2 / 2 * dy * dz
    )


test_integrate_gauss_cuboid()
```
